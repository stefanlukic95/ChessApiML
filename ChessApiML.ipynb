```python
import pickle
from skimage.io import imread
from skimage.transform import resize
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
import pathlib
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath
```

```python
test_model = pickle.load(open("Chess_model.p","rb"))
```

```python
Categories = ["Knight","Queen","Rook", "Pawn", "King", "Bishop"]
```

```python
# Testing on new image
flat_data = []
url = input ('Enter url of image to test: ')
img_array = imread(url)

img_resized = resize(img_array,(150,150,3))
flat_data.append(img_resized.flatten())
flat_data = np.array(flat_data)
print("Dimensions of original image are:",img_array.shape)
plt.imshow(img_resized)
y_output = test_model.predict(flat_data)
y_output = Categories[y_output[0]]
#Test example URLs:
#https://upload.wikimedia.org/wikipedia/commons/thumb/a/af/Chess_piece_-_White_queen.jpg/131px-Chess_piece_-_White_queen.jpg
#https://upload.wikimedia.org/wikipedia/commons/thumb/7/7e/Chess_piece_-_White_king.jpg/131px-Chess_piece_-_White_king.jpg
#https://upload.wikimedia.org/wikipedia/commons/thumb/b/ba/Chess_piece_-_Black_bishop.JPG/800px-Chess_piece_-_Black_bishop.JPG
#https://media.istockphoto.com/vectors/silhouette-of-a-knight-chess-piece-vector-id1143551796
#https://upload.wikimedia.org/wikipedia/commons/thumb/f/fe/Chess_piece_-_White_rook.JPG/800px-Chess_piece_-_White_rook.JPG
#https://upload.wikimedia.org/wikipedia/commons/thumb/f/ff/Chess_piece_-_Black_knight.JPG/800px-Chess_piece_-_Black_knight.JPG
#https://www.ubergames.co.uk/wp-content/uploads/2021/05/uber-games-chess-pieces-pawn-black-6-ug350.jpg
print("Predicted:",y_output)
```
